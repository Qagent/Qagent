```python
import torch
from torch.utils.data import Dataset

class CustomDataset(Dataset):
    def __init__(self, annotations_file, data, shuffle=None, scramble=None):
        self.labels = annotations_file
        self.data = data

    def __len__(self):
        return int(self.data.size/2)

    def __getitem__(self, idx):
        return self.data.iloc[idx,1]

HindiCSV  =  CustomDataset(chars, df)
H = {}
for i in range(len(HindiCSV)):
  try:
    lit = HindiCSV[i].split(" ")
  except:
    print(i)

  for j in lit:
    if j in H: H[j] += 1
    else: H[j] = 1
```
